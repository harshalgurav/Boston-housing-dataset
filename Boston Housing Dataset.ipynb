<b><span style=font-size:20px>- <font color='MediumBlue'>Importing Libraries</font></span></b>

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Separating Independent and Dependent variable to Train and Test the Dataset

X = df.drop('MEDV',axis=1)
Y = df['MEDV']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,random_state=1)
```

<b><span style=font-size:20px>- <font color='MediumBlue'>Appling StandardScaler</font></span></b>

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

<center><b><span style=font-size:25px><font color='MediumBlue'><u>Linear Regression</u></font></span></b></center>

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
# Checking the Slope and Intercept of multiple inputs after Training.

m = lr.coef_
print('Slope : ', m)

c = lr.intercept_
print('Intercept : ', c)
```

```python
# Now that we have Trained the data, we need to Test it.

Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

# Calculating Root Mean Squared Error (RMSE) = sqrt(MSE) to reduce error.

mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)

print('Root Mean Squared Error :', rmse)
print('Mean Squared Error :', mse)
```

```python
print("Model Score: ",lr.score(X_test, Y_test))
```

<b><span style=font-size:20px>- <font color='LimeGreen'>Conclusion.</font></span></b>
1. Here, we got 76% in Linear Regression 
2. But we will still try to get better accuracy
3. So, we will check if the dataset satisfies all the Linear Regression Assumptions or not.

<center><b><span style=font-size:25px><font color='MediumBlue'><u>Assumptions of Linear Regression</u></font></span></b></center>

<b><span style=font-size:15px>1. <font color='OrangeRed'>Multicollinearity</font></span></b>

```python
# plotting a Heatmap for checking the Correlation between inputs.

plt.figure(figsize = (20,8))
sns.heatmap(df.corr(), annot = True)
plt.show()


# Selecting features which have High Correlation

plt.figure(figsize=(20, 5))

features = ['LSTAT','PTRATIO','RM']
target = df['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    X = df[col]
    Y = target
    plt.scatter(X,Y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

<b><span style=font-size:20px>- <font color='LimeGreen'>Conclusion.</font></span></b>
1. From above Heat Map we have selected features
    - LSTAT
    - PTRATIO
    - RM
2. From above graph 'PTRATIO' do not have Linear Relationship so we will eliminate it.
3. So, we will only select 'RM' & 'LSTAT' column as Input

```python
X = df[['RM','LSTAT']]
Y = df['MEDV']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,random_state=1)
```

<b><span style=font-size:20px>- <font color='MediumBlue'>Appling StandardScaler</font></span></b>

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

<center><b><span style=font-size:25px><font color='MediumBlue'><u>Linear Regression</u></font></span></b></center>

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
# Checking the Slope and Intercept of multiple inputs after Training.

m = lr.coef_
print('Slope : ', m)

c = lr.intercept_
print('Intercept : ', c)
```

```python
# Now that we have Trained the data, we need to Test it.

Y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

# Calculating Root Mean Squared Error (RMSE) = sqrt(MSE) to reduce error.

mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)

print('Root Mean Squared Error :', rmse)
print('Mean Squared Error :', mse)
```

```python
print("Model Score: ",lr.score(X_test, Y_test))
```

<b><span style=font-size:20px>- <font color='LimeGreen'>Conclusion.</font></span></b>
1. Linear Regression: 76%
2. After applying Multicollinearity: 68% 

```python
residual = Y_test - Y_pred
print(residual)
```

<b><span style=font-size:15px>2 & 3. <font color='OrangeRed'>Linear Relationship & 'Homoskedastik' or 'Heteroskedastik'.</font></span></b>

```python
sns.scatterplot(Y_pred,residual)
plt.show()
```

<b><span style=font-size:15px>4. <font color='OrangeRed'>Residual Curve</font></span></b>

```python
# Plotting a Histplot() for Resiudal.

sns.histplot(residual)
plt.show()

# Plotting a Distplot() for Resiudal.

sns.distplot(residual)
plt.show()

print('Skewness for Residual : ', residual.skew())
```

<b><span style=font-size:20px>- <font color='LimeGreen'>Conclusion.</font></span></b>
1. The plot graph came Heteroskedastik.
2. It did not showed Linear Relation.
3. Residual skewness did not came properly.
4. Only Multicollinearity assumption got satisfied i.e. 1/4 assumption got satisfied.

<center><b><span style=font-size:25px><font color='MediumBlue'><u>Polynomial Regression</u></font></span></b></center>

```python
X = df[['RM','LSTAT']]
Y = df['MEDV']
```

```python
# Calling the inbuilt class PolynomialFeatures from file preprocessing which is inside sklearn folder.

from sklearn.preprocessing import PolynomialFeatures
```

```python
# Increasing the degree of input x using PolynomialFeatures class.

pf = PolynomialFeatures(3)
```

```python
X_poly = pf.fit_transform(X)
```

```python
# Splitting the data for Training and Testing purposes in 70:30 ratio using x_poly which as degree '2'

X_train,X_test,Y_train,Y_test = train_test_split(X_poly,Y,test_size = 0.3,random_state = 1)
```

```python
lr.fit(X_train, Y_train)
```

```python
Y_pred = lr.predict(X_test)
```

```python
# Finding Slope and Intercept.

m = lr.coef_
print('Slope : ', m)

c = lr.intercept_
print('Intercept : ', c)
```

```python
# Calculating Mean Squared Error and Root Mean Squared Error.

mse = mean_squared_error(Y_test,Y_pred)
print('Mean Squared Error : ', mse)

rmse = np.sqrt(mse)
print('Root Mean Squared Error : ', rmse)
```

```python
print("Model Score: ",lr.score(X_test, Y_test))
```
